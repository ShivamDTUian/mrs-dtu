```python
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=4:
            L.append(i['name'])
            counter += 1 
        else: 
                break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
#here aslo we extract top 4 name from the crew as well whose having job value as director only so similarly we use above convert fn
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
##here are the top 4 names having jobs as a director
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer 
ps  = PorterStemmer()
```

```python
def stem(text):
    y = [] 
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']  = new_df['tags'].apply(stem)
```

```python
#here we fetching the index for calc dist for each movie and to recommend that movie with similar one
def recommend(movie):
    movie=movie.lower()
    try:
        movie_index = int(new_df[new_df['title'] == movie].index[0]) 
    except IndexError: # This will handle the case if the movie is not found
        print(f"Movie '{movie}' not found in the database.")  #concept Fstring
        return 
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:11] #using dist instead of similarity
    for i in movie_list :
        print((new_df.iloc[i[0]].title).upper())
```

```python
recommend('Avatar')
```
